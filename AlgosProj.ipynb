```python
import pandas as pd
import numpy as np

data = pd.read_csv('movies.csv')
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data = data.drop(columns = ['color','director_name', 'actor_2_name', 'genres', 'actor_1_name', 'movie_title', 'actor_3_name', 'plot_keywords'
                           ,'movie_imdb_link','language', 'country','content_rating', 'title_year','facenumber_in_poster','aspect_ratio'])
data.head()
```

```python
data.shape
```

```python
data[data['director_facebook_likes'] < 1]
```

```python
data2 = data.replace(0, np.NaN)
data2.head()
```

```python
data2.director_facebook_likes.mean()
```

```python
data3 = data2.director_facebook_likes.fillna((data2.director_facebook_likes.mean()/10)* data2.imdb_score)
```

```python
data2.director_facebook_likes = data3
data2.head()
```

```python
data3 = data2.num_critic_for_reviews.fillna((data2.num_critic_for_reviews.mean()/10)* data2.imdb_score)
data2.num_critic_for_reviews = data3

data2.num_critic_for_reviews.mean()
```

```python
data3 = data2.duration.fillna(data2.duration.mean())
data2.duration = data3

data2.head()
```

```python
data3 = data2.actor_3_facebook_likes.fillna((data2.actor_3_facebook_likes.mean()/10) *data2.imdb_score)
data2.actor_3_facebook_likes = data3

data2.head()
```

```python
data2[data2['budget'] == np.NaN]
```

```python
data2[data2['movie_facebook_likes'].isna()].shape
```

```python
data2 = data2.drop(columns = ['movie_facebook_likes'])
```

```python
data3 = data2.gross.fillna((data2.gross.mean()/10* data2.imdb_score))
data2.gross = data3
```

```python
data2[data2['gross'].isna()].shape
```

```python
data2.head()
```

```python
data3 = data2.budget.fillna((data2.budget.mean()/10* data2.imdb_score))
data2.budget = data3

data2.head()
```

```python
data2.shape
```

```python
data2 = data2.dropna()
```

```python
data2.columns
```

```python
from sklearn import svm

trainedSet = data2

trainedSet['budget'] = trainedSet['budget']/1000
trainedSet['gross'] = trainedSet['gross']/1000

trainedSet = trainedSet.astype({'director_facebook_likes':'int64'})
trainedSet = trainedSet.astype({'actor_3_facebook_likes':'int64'})
trainedSet = trainedSet.astype({'actor_1_facebook_likes':'int64'})
trainedSet = trainedSet.astype({'duration':'int64'})
trainedSet = trainedSet.astype({'num_critic_for_reviews':'int64'})
trainedSet = trainedSet.astype({'actor_2_facebook_likes':'int64'})
trainedSet = trainedSet.astype({'cast_total_facebook_likes':'int64'})
trainedSet = trainedSet.astype({'num_user_for_reviews':'int64'})
```

```python
trainedSet
```

```python
trainedSet.dtypes
```

```python
y = trainedSet['imdb_score']
X = trainedSet.drop(columns=['imdb_score'])

X = np.asarray(X)
y = np.asarray(y)
X[0:10]
```

```python
from sklearn.svm import SVR
import numpy as np


clf = SVR(C=1.0, epsilon=0.01, gamma='auto')
clf.fit(X, y)
#scores = cross_val_score(clf, X_train, y, cv=5)
#scores
```
